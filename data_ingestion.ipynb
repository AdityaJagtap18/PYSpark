```python
import os
import pandas as pd

out_dir = "data/hnm"
os.makedirs(out_dir, exist_ok=True)

files = {
    "articles.csv": "hf://datasets/einrafh/hnm-fashion-recommendations-data/data/raw/articles.csv",
    "customers.csv": "hf://datasets/einrafh/hnm-fashion-recommendations-data/data/raw/customers.csv",
    "transactions_train.csv": "hf://datasets/einrafh/hnm-fashion-recommendations-data/data/raw/transactions_train.csv",
}

for name, hf_path in files.items():
    print(f"Downloading {name} ...")
    df = pd.read_csv(hf_path)
    df.to_csv(os.path.join(out_dir, name), index=False)
    print(f"Saved -> {os.path.join(out_dir, name)} | shape={df.shape}")

print("Done.")
```
